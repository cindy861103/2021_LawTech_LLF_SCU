```python
import plotly
import plotly.graph_objs as go
```

```python
import re
import pandas as pd

data = pd.read_csv("drunkDrive_2018.csv",sep="\t")

data.head()
```

```python
s = list(set(data["不能安全駕駛有期徒刑_文字分組"]))
print(s)
#唯一值 判刑區間只有這些
```

```python
a = list(set(data["酒精濃度"]))

#唯一值 出現過的酒精濃度只有這些

alc=[]
for g in a:
    g=str(g)
    if g!="nan":
        alc.append(float(g))
print(alc)

bac=["0.0~0.4","0.5~1","1.1~1.5","1.6~2","2.1~2.5"]
```

```python
judge = list(data["不能安全駕駛有期徒刑_文字分組"])
judge_count= []

for i in s:
    print(i,":",judge.count(i))
    judge_count.append(judge.count(i))

        
#2018年每種刑度有幾筆
```

```python
judge_count
#2018年每種刑度有幾筆
```

```python
plotly.offline.init_notebook_mode(connected=True)

plotly.offline.iplot({
    "data": [go.Scatter(x=s, y=judge_count)],
    "layout": go.Layout(title="刑度區間及其被判次數關係圖")
})
```

```python
d= list(set(data["車種"]))
print(d)
```

```python
judge = list(data["車種"])
judge_count= []

for i in d:
    print(i,":",judge.count(i))
    judge_count.append(judge.count(i))
    
#2019年每種刑度有幾筆
```
