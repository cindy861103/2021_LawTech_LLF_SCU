```python
wine = input("請輸入喝的酒種：（紅酒/啤酒/米酒/米酒頭/白酒/高粱酒/大麴酒/紹興酒）")
shot = int(input("請輸入喝酒的杯數"))
```

```python
alcohol = {"紅酒":0.0852,
           "啤酒":0.0390,
           "紹興酒":0.1250,
           "米酒":0.1560,
           "米酒頭":0.2737,
           "白酒":0.3125,
           "高粱酒":0.4518,
           "大興酒":0.5068}
```

```python
target = alcohol[wine] * shot
print("初犯：吐氣濃度酒測值為：",target)
```

```python
target_1=alcohol[wine] * shot*you
print("累犯：吐氣濃度酒測值為：",target_1)
```
