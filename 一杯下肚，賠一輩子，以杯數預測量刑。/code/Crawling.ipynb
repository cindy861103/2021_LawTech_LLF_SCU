```python
from time import sleep        # 爬蟲禮貌，避免對方伺服器無法負荷請求而宕機
import requests               # 爬蟲請求
import json                   # JSON 資料型態
import pandas as pd  # 資料處理套件
from bs4 import BeautifulSoup

first_clean_list=[]
number=[]
date=[]
reason=[]
content=[]

for i in range(1,4629):
    #查找網站
    try:
        q = str(i-1).format()
        res = requests.get(f'https://law.judicial.gov.tw/FJUD/data.aspx?ro={q}&q=ccd70cfd945a6e127c9d17caf5f77dce&sort=DS&ot=in')
        soup = BeautifulSoup(res.text , 'html.parser')

        target_table = soup.find_all(class_='col-td')
        print(target_table[0])
        number.append(target_table[0].text)
        date.append(target_table[1].text)
        reason.append(target_table[2].text)
        content.append(target_table[3].text)
    except: pass
    sleep(0.5)

    
apple_array=[]
for i in range(len(number)):
    apple_array.append({
                            "裁判字號": number[i],
                            "裁判日期": date[i],
                            "裁判案由":reason[i],
                            "內文":content[i]})
apple_df = pd.DataFrame.from_dict(apple_array)
```
